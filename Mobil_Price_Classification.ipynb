## Preprocessing

```python
x=df_train.drop('price_range',axis=1)
y=df_train['price_range']
```

```python
scaler=StandardScaler()
x_scaled=scaler.fit_transform(x)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.2,random_state=42)
```

## Modeling

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression

model_lg=LogisticRegression()
scores=cross_val_score(model_lg,x_train,y_train,cv=5,scoring='r2')

print("R2 scores for each fold:",scores)
print("Average R2:", scores.mean())
```

```python
model_lg.fit(x_train,y_train)
```

```python
y_pred=model_lg.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
from sklearn.ensemble import RandomForestClassifier

model_Rf=RandomForestClassifier()
scores = cross_val_score(model_Rf,x, y, cv=5, scoring='r2') 

print("R2 scores:", scores)
print("Average R2:", scores.mean())
```

```python
model_Rf.fit(x_train,y_train)
```

```python
y_pred_rf=model_Rf.predict(x_test)
```

```python
accuracy_score(y_test,y_pred_rf)
```

```python
from sklearn.svm import SVC

model_svc=SVC()
scores = cross_val_score(model_svc,x, y, cv=5, scoring='r2') 

print("R2 scores:", scores)
print("Average R2:", scores.mean())
```

```python
model_svc.fit(x_train,y_train)
```

```python
y_perd_svc=model_svc.predict(x_test)
```

```python
accuracy_score(y_test,y_perd_svc)
```

```python
from xgboost import XGBClassifier

model_xgb=XGBClassifier()
scores = cross_val_score(model_xgb, x_train, y_train, cv=5, scoring='r2')

print("R2 scores for each fold:", scores)
print("Average R2:", scores.mean())
```

```python
model_xgb.fit(x_train, y_train)
```

```python
y_pred_xgb = model_xgb.predict(x_test)
```

```python
accuracy_score(y_test, y_pred_xgb)
```

```python
from sklearn.tree import DecisionTreeClassifier

model_dt = DecisionTreeClassifier()
scores = cross_val_score(model_dt, x_train, y_train, cv=5, scoring='r2')

print("R2 scores for each fold:", scores)
print("Average R2:", scores.mean())
```

```python
model_dt.fit(x_train, y_train)
```

```python
y_pred_decision = model_dt.predict(x_test)
```

```python
accuracy_score(y_test, y_pred_decision)
```
